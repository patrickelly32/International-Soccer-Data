```python
# Import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
#import geopy
import requests
import json
#import citipy
```

```python
# Read CSV File
df1 = pd.read_csv('Updated_cleaned_DF.csv')
df1.head()

#Group by the year and find the average goals made 
group_year = df1.groupby('date')
totalGoalDF = pd.DataFrame(group_year['total goals'].mean())

# Set X and Y values to plot
x_axis = totalGoalDF.index
y_val = totalGoalDF['total goals']


#Add titles and format plot
plt.title("Goals Per Game: 1872-2019")
plt.xlabel("Years")
plt.ylabel("Goals Per Game")
plt.grid()
plt.scatter(x_axis, y_val)
plt.savefig("goals_per_game.png")
```
